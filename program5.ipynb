```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
x=np.random.rand(100,1)
lables=np.array([1 if xi<=0.5 else 2 for xi in x[:50]])
x_train , y_train = x[:50], lables
x_test = x[:50]
k_values = [1, 2, 3, 4, 5, 20, 30]
plt.figure(figsize=(12,8))
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    plt.scatter(x_test, y_pred, label=f'k={k}', alpha=0.7)
plt.scatter(x_train, y_train, color='red', marker='x', label='Training Data')
plt.xlabel('X values')
plt.ylabel('Predicted Class')
plt.title('KNN Classification for Different k-values')
plt.legend()
plt.show()

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    print(f'Predictions for k={k}:', y_pred)

```
